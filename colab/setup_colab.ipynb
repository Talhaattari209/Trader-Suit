# Colab Setup — GPU, Repo, Dependencies

Run this notebook first. It will:
1. Check that GPU is enabled (Runtime → Change runtime type → GPU).
2. Clone the repo or use an existing path.
3. Install `requirements-colab.txt`.
4. Add the project root to `sys.path` so you can import `src` modules.

```python
# 1) Check GPU — Runtime → Change runtime type → GPU
try:
    import torch
    device = "cuda" if torch.cuda.is_available() else "cpu"
    print(f"PyTorch device: {device}")
    if device == "cuda":
        print(f"  GPU: {torch.cuda.get_device_name(0)}")
except Exception as e:
    print("PyTorch not yet installed; run the pip cell below first.")
    print(str(e))
```

```python
# 2) Set repo path: clone from GitHub OR use uploaded/unzipped path
import os

REPO_URL = "https://github.com/YOUR_USER/claude.git"  # Change to your repo
PROJECT_ROOT = "/content/claude"  # Where the project lives after clone/unzip

if not os.path.exists(PROJECT_ROOT):
    print(f"Cloning into {PROJECT_ROOT}...")
    os.makedirs("/content", exist_ok=True)
    import subprocess
    subprocess.check_call(["git", "clone", "--depth", "1", REPO_URL, PROJECT_ROOT])
else:
    print(f"Using existing project at {PROJECT_ROOT}")

os.chdir(PROJECT_ROOT)
print(f"Project root: {os.path.abspath(os.curdir)}")
```

```python
# 3) Install Colab requirements (run once)
import subprocess
COLAB_REQ = os.path.join(PROJECT_ROOT, "colab", "requirements-colab.txt")
if os.path.exists(COLAB_REQ):
    subprocess.check_call(["pip", "install", "-q", "-r", COLAB_REQ])
    print("Installed requirements-colab.txt")
else:
    subprocess.check_call(["pip", "install", "-q", "torch", "stable-baselines3", "gym", "scikit-learn", "pandas", "numpy"])
    print("Installed core packages (colab/requirements-colab.txt not found)")
```

```python
# 4) Add project root to Python path
import sys
if PROJECT_ROOT not in sys.path:
    sys.path.insert(0, PROJECT_ROOT)
print("sys.path includes:", PROJECT_ROOT)
```
